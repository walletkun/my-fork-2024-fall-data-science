### Clean the Data

Perhaps we’d like to know more another day, but the team is really interested in just the relationship between calories (`calories_day`) and weight. …and maybe gender.

Can you remove the other columns?

```python
# Remove ‘em.
relationship_of_columns = df_food[['calories_day','weight','Gender']]
print(relationship_of_columns)


#filtering out all the alphanumeric value only keeping the characters
def filter_alpha(value):
    if isinstance(value, str):
        result = ''.join(re.findall(r'[^a-zA-Z0-9]',value))
        return result if result else np.nan
    return np.nan


is_string = df_food['weight'].apply(filter_alpha)

print(is_string)
```

What about `NaN`s? How many are there?

```python
# Count ‘em.
#We can use isna function from pandas
count_nan = df_food.isna().sum().sum()
#The first sum is going to aggregate the values into boolean values, where 1 is true 0 is false
#Second sum is going to sum all those values together
print(count_nan)
```

We gotta remove those `NaN`s―the entire row.

```python
# Drop ‘em.
#We can drop all of those NaNs by using dropna
dropped_nan = df_food.dropna()
print(dropped_nan)
```

But what about the weird non-numeric values in the column obviously meant for numeric data?

Notice the data type of that column from when you got the types of all the columns?

If only we could convert the column to a numeric type and drop the rows with invalid values. 🤔

```python
# Fix that.
#First we can use .to_numeric() to handle each columns into numeric values (that are necessary) and give all those non-numeric a NaN 
df_food['GPA'] = pd.to_numeric(df_food['GPA'], errors='coerce')
df_food['weight'] = pd.to_numeric(df_food['weight'], errors='coerce')
df_food['weight'].dropna()
df_food['GPA'].dropna()
```

Now this data seems reasonably clean for our purposes! 😁
